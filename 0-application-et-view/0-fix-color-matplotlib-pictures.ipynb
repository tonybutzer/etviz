```python
import xarray as xr
def img_plot_proxy(variable='dd_', collection='s3://dev-et-data/enduser/DelawareRiverBasin/drb150/Annual/', 
                   year=2018, month=10):
    #img = 's3://dev-et-data/enduser/DelawareRiverBasin/drb150/Annual/'
    img=collection
    img = img + variable + str(year) + '.tif'
    da = xr.open_rasterio(img)
    #da_masked = da.where(da>0.0)
    da_masked = da.where(da<10000.0)
    da_masked = da_masked.where(da>0.0)
    img_figure = da_masked.plot().get_figure()
    img_figure = da_masked.plot
    #print(help(da_masked.plot()))
    return(img_figure)
```
