```python
# Implementation of matplotlib function 
import matplotlib.pyplot as plt 
import numpy as np 
t = np.arange(10) 
   
  
plt.plot(t, np.sin(t)) 
     
plt.title('matplotlib.pyplot.ginput()function Example', fontweight ="bold") 
   
#print("After 3 clicks :") 
#x = plt.ginput(3) 
plt.ginput(n=1, timeout=30, show_clicks=True,
       mouse_add=1, mouse_pop=3, mouse_stop=2)
#print(x) 
   
plt.show() 

x = plt.ginput(n=1, timeout=30, show_clicks=True,
       mouse_add=1, mouse_pop=3, mouse_stop=2)

print(x)
```
